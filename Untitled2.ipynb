```python
import scipy.stats as stats
variables = [summary_table]
# Group the 'data' DataFrame by 'sexo', calculate mean and standard deviation
grouped_data = data.groupby('sexo').agg(['mean', 'std'])

# Reorganiza el DataFrame para que las variables sean filas y las estadísticas columnas
grouped_data = grouped_data.stack(level=0).reset_index().rename(columns={'level_1': 'variable'})
grouped_data = grouped_data.pivot(index='variable', columns='sexo', values=['mean', 'std'])

# Ordena las filas según la lista 'orden_variables'
grouped_data = grouped_data.loc[orden_variables]
p_values = []
for variable in grouped_data.index:
    grupo_hombre = data[data['sexo'] == '1'][variable]
    grupo_mujer = data[data['sexo'] == '0'][variable]

    # Selecciona la prueba adecuada según la distribución de los datos
    if stats.shapiro(grupo_hombre)[1] > 0.05 and stats.shapiro(grupo_mujer)[1] > 0.05:
        # Prueba t de Student (datos normales)
        p_value = stats.ttest_ind(grupo_hombre, grupo_mujer)[1]
    else:
        # Prueba U de Mann-Whitney (datos no normales)
        p_value = stats.mannwhitneyu(grupo_hombre, grupo_mujer)[1]

    p_values.append(p_value)

# Agrega la columna de valores p y la indicación
grouped_data['p-value'] = p_values
grouped_data['p < 0.05'] = grouped_data['p-value'] < 0.05

variables.append(grouped_data)
print(grouped_data)
```

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for plotting

# Assuming 'data' is your DataFrame containing the variables
variables = ['imc','laps', 'potmedia', 'ppeak', 'l1', 'l2', 'l3']
subset_data = data[variables]

# Calculate the correlation matrix
correlation_matrix = subset_data.corr()

# Calculate p-values for correlations
p_values = []
for i in range(len(correlation_matrix.columns)):
    row = []
    for j in range(len(correlation_matrix.columns)):
        x = subset_data[correlation_matrix.columns[i]].replace([np.inf, -np.inf], np.nan)
        y = subset_data[correlation_matrix.columns[j]].replace([np.inf, -np.inf], np.nan)

        # Drop NaN values from both arrays simultaneously to ensure equal length
        common_indices = x.index.intersection(y.index)
        x_clean = x.loc[common_indices].dropna()
        y_clean = y.loc[common_indices].dropna()

        # Check if there's enough data after removing NaNs and infs
        if len(x_clean) > 1 and len(y_clean) > 1:
            # Recheck lengths before calling pearsonr
            if len(x_clean) != len(y_clean):
                print(f"Warning: Length mismatch for {correlation_matrix.columns[i]} and {correlation_matrix.columns[j]}")
                row.append(None)
            else:
                _, p_value = stats.pearsonr(x_clean, y_clean)
                row.append(p_value)
        else:
            row.append(None)  # Indicate insufficient data
    p_values.append(row)

p_value_matrix = pd.DataFrame(p_values, columns=correlation_matrix.columns, index=correlation_matrix.columns)

# Create a heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

# Pairplot to visualize relationships between variables
sns.pairplot(subset_data)
plt.show()

print("Matriz de Correlación:")
print(correlation_matrix)
print("\nMatriz de Valores p:")
print(p_value_matrix)
```

```python
# prompt: para realizar analisis de regresion lineal multiple jerarquizada que incluya estas variables; y para explore the model by examining coefficients, p-values, and other statistics.

import statsmodels.formula.api as smf


# Assuming 'data' is your DataFrame

# Define the hierarchical model
model_formula = "l1 ~ laps + zimc + z score Potmedia + l1 "

# Fit the model
model = smf.ols(formula=model_formula, data=data).fit()

# Print the model summary
print(model.summary())

# Access coefficients
coefficients = model.params
print("Coefficients:\n", coefficients)

# Access p-values
p_values = model.pvalues
print("\nP-values:\n", p_values)

# Access R-squared
r_squared = model.rsquared
print("\nR-squared:", r_squared)

# Access adjusted R-squared
adj_r_squared = model.rsquared_adj
print("\nAdjusted R-squared:", adj_r_squared)

# Access F-statistic
f_statistic = model.fvalue
print("\nF-statistic:", f_statistic)

# Access p-value of F-statistic
f_p_value = model.f_pvalue
print("\nP-value of F-statistic:", f_p_value)

```
